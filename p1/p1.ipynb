```python
def smart_count(nums):
    total = 0
    encounter_neg = False
    for x in nums:
        if x < 0:
            encounter_neg = True
        if x <= 10 and not encounter_neg:
            total+=x
    return total
```

```python
import os, sys
import pandas
filenames = os.listdir("home-computers")
years = dict()
for name in filenames:
    if name[:5]=="ACSDT":
        years[int(name[7:11])] = pandas.read_csv(os.path.join("home-computers", name),skiprows=1)
```

```python
#Q17: what are the keys in years?
sorted(list(years.keys()))
```

```python
#Q18: how many households did Wisconsin have in 2018?
df = years[2018]
# df.loc[df["Geographic Area Name"]=="Wisconsin"].at["Estimate!!Total"]
df.loc[df["Geographic Area Name"]=="Wisconsin","Estimate!!Total"].values[0]
```

```python
df
```

```python
#Q19: how many total households in the US are estimated to not have any computer at home? (2018)
sum(df["Estimate!!Total!!No Computer"])
```

```python
#Q20: what is the biggest per-state margin of error for "No Computer", as a fraction of the total estimate? (2018)
(df["Margin of Error!!Total!!No Computer"]/df["Estimate!!Total!!No Computer"]).max()
```

```python
states=["Illinois", "Indiana", "Iowa", "Michigan", "Minnesota", "Wisconsin"]
q21_ans = dict()
for state in states:
    total = df.loc[df["Geographic Area Name"]== state,"Estimate!!Total"].values[0]
    num_no_computer = df.loc[df["Geographic Area Name"]== state,"Estimate!!Total!!No Computer"].values[0]
    q21_ans[state] = total/num_no_computer

q21_ans 
```

```python
#Q22: same question, but answer with a bar plot.
p22 = pandas.Series(q21_ans).plot.bar()
p22.set_xlabel("Geographic Area Name")
p22.set_ylabel("% of Households w/o Computers")
p22
```

```python
#Q23: how as the number of WI households without computers changed over recent years?
q23_dict = dict()
for year in sorted(list(years.keys())):
    df = years[year]
    num = df.loc[df["Geographic Area Name"]== "Wisconsin","Estimate!!Total!!No Computer"].values[0]
    q23_dict[year] = num/1000
    
p23 = pandas.Series(q23_dict).plot.line()
p23.set_xlabel("Year")
p23.set_ylabel("WI Households without Computers (thousands)")
p23
```

```python
#Q24: what is the relationship between household with smartphones and those with tablets?
q24_dict = dict()
phones = []
tablets = []
for year in sorted(list(years.keys())):
    df = years[year]
    if "Estimate!!Total!!Has one or more types of computing devices!!Smartphone" in df.columns and "Estimate!!Total!!Has one or more types of computing devices!!Tablet or other portable wireless computer" in df.columns:
        phones += list(df["Estimate!!Total!!Has one or more types of computing devices!!Smartphone"])
        tablets+=list(df["Estimate!!Total!!Has one or more types of computing devices!!Tablet or other portable wireless computer"])
s1 = pandas.Series(phones)/1000000
s2 = pandas.Series(tablets)/1000000
df = pandas.DataFrame({
    "Smartphones(millions)": s1,
    "Tablets(millions)": s2
})
df.plot.scatter(x = "Smartphones(millions)",y="Tablets(millions)")
```

```python
from subprocess import check_output
```

```python
#Q25: what is the first line of output from git log?
os.chdir("./cs320-p1")
output = check_output(["git", "log"],cwd = ".")
str_output = str(output, encoding="utf-8")
log_lst = str_output.split("\n")
log_lst[0]
```

```python
#Q26: What are all the commit numbers?
q26_ans = []
for i in range(0,len(log_lst),6):
    q26_ans.append(log_lst[i].split(" ")[1])
q26_ans
```

```python
#Q27: how has the length of wc.py changed over time?
q27_ans = {}
for num in q26_ans:
    check_output(["git", "checkout", num],cwd = ".")
    f = open("./wc.py")
    lines = list(f)
    f.close()
    q27_ans[num] = len(lines)
check_output(["git", "checkout", "main"],cwd = ".")
q27_ans
```

```python
#Q28: what does python3 cs320-p1/wc.py input.txt ALL return when input.txt contains "A B C C"?
os.chdir("..")
f = open("./input.txt", "w")
f.write("A B C C")
f.close()
data = json.loads(check_output(["python3", "cs320-p1/wc.py", "input.txt", "ALL"],cwd = "."))
print(data)
```

```python
#Q29: for which versions (commits) of the program does the previous command fail?
error_versions = ""
for num in q26_ans:
    os.chdir("./cs320-p1")
    check_output(["git", "checkout", num],cwd = ".")
    os.chdir("..")
    try:
        data = json.loads(check_output(["python3", "cs320-p1/wc.py", "input.txt", "ALL"],cwd = "."))
    except:
        error_versions = num
check_output(["git", "checkout", "main"],cwd = ".")
error_versions
```

```python
#Q30: for commit 4e4128313b8d5b5e5d04f2e8e585f64f7c5831a4, what is the relationship between input size and time to run wc.py?
import time
word_sizes = [200,400,600,800,1000]
relationships = dict()
check_output(["git", "checkout", "main"],cwd = ".")

if not os.getcwd()[-8:] == 'cs320-p1':
    os.chdir("./cs320-p1")
check_output(["git", "checkout", "4e4128313b8d5b5e5d04f2e8e585f64f7c5831a4"],cwd = ".")
os.chdir("..")
  
    
for wsize in word_sizes:
    f = open("./input.txt", "w")
    for i in range(wsize):     
        f.write(str(i))
    f.close()
    
    start_time = time.time()
    check_output(["python3", "cs320-p1/wc.py", "input.txt", "ALL"],cwd = ".")
    relationships[wsize] = (time.time() - start_time)*1000
    
p30 = pandas.Series(relationships).plot.line()
p30.set_xlabel("Worlds")
p30.set_ylabel("Milliseconds")
p30
```
